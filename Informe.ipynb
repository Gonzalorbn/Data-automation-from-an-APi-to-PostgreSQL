# **AUTOMATIZACIÓN DE DATOS DESDE UNA API A POSTGRESQL**

## **PROPÓSITO DEL PROYECTO**

Este proyecto tiene como objetivo principal automatizar el proceso de adquisición de datos desde la API de Mercado Libre, su almacenamiento eficiente en una base de datos PostgreSQL, y permitir a los usuarios realizar consultas personalizadas directamente desde la terminal si así lo desean. Los objetivos específicos incluyen:



*   Automatización de la Adquisición de Datos: Implementar un sistema que consulte la API de Mercado Libre de manera periódica para obtener datos actualizados relacionados con productos en línea.


*   Almacenamiento Eficiente: Crear una estructura de base de datos en PostgreSQL que permita el almacenamiento seguro y organizado de los datos recopilados de la API.

*   Acceso a Datos Relevantes: Proporcionar a los usuarios la capacidad de realizar consultas SQL personalizadas en la base de datos directamente desde la terminal, lo que les permitirá acceder a información específica de interés de manera eficiente.


*   Simplificación de Procesos: Simplificar la gestión de conexiones y transacciones de la base de datos a través de un manejo eficiente de cursores y conexiones.


*   Registro de Eventos: Implementar un registro de eventos que facilite la identificación de problemas y errores, así como un seguimiento de las operaciones realizadas.
*   Flexibilidad y Escalabilidad: Diseñar el sistema de manera que sea flexible y permita futuras expansiones y adaptaciones para abordar nuevas fuentes de datos o requisitos adicionales.






## **DETALLES DEL PROYECTO**


### TECNOLOGÍAS UTILIZADAS


*   Visual Studio Code (VSCode): Utilizado como entorno de desarrollo para escribir, depurar y administrar código.

*   Postman: Herramienta utilizada para probar y depurar APIs.

*   PostgreSQL: Base de datos utilizada para almacenar y gestionar los datos adquiridos de la API de Mercado Libre.



*   DBeaver: Herramienta de administración de bases de datos de código abierto que se puede utilizar para conectarse a bases de datos PostgreSQL.











### DESCRIPCIÓN BREVE DE LOS SCRIPTS UTILIZADOS


*   conexion.py:
Define una clase llamada Conexion que se encarga de gestionar conexiones a una base de datos PostgreSQL. Utiliza la biblioteca psycopg2 para crear un pool de conexiones que facilita la gestión de conexiones y cursores de la base de datos. Es esencial para el acceso a la base de datos y garantiza la disponibilidad de conexiones en el pool.

*   consult_api.py:
Este script contiene la clase Consulta_API, que se encarga de realizar consultas a la API de Mercado Libre y guardar los datos relevantes en un archivo TSV. Utiliza la biblioteca requests para realizar las solicitudes a la API y asegura que los datos se almacenan en un formato legible.


*   crear_tabla.py:
Contiene la clase GestorBaseDatos, que se utiliza para crear una tabla en la base de datos PostgreSQL. Esta tabla se crea con una estructura específica que corresponde a los datos obtenidos de la API. El script también incluye la funcionalidad para copiar los datos desde un archivo TSV a la tabla recién creada.

*   cursor_del_pool.py:
Este script define la clase CursorDelPool, que se utiliza para gestionar conexiones y cursores de la base de datos en un contexto with. Facilita la obtención de conexiones desde el pool, la gestión de transacciones y la liberación de recursos. Garantiza que las operaciones de base de datos se realicen de manera segura.


*   logger_base.py:
Configura el registro de eventos (logging) para el proyecto. Establece los niveles de registro, el formato de los mensajes y los controladores (handlers) para la salida de registro en archivos y en la consola. Ayuda a realizar un seguimiento de eventos y errores en la aplicación.

*   consultas.py:
El script consultas.py contiene la clase Consultas, que se encarga de realizar consultas SQL personalizadas en la base de datos PostgreSQL. Utiliza la clase CursorDelPool para gestionar las conexiones y los cursores de la base de datos. Facilita la interacción con la base de datos para recuperar datos de acuerdo con las consultas del usuario.


*   ejecutar_consultas.py:
Es el programa principal que interactúa con el usuario. Proporciona un menú para realizar consultas personalizadas en la base de datos y muestra los resultados. Además, se encarga de la extracción de datos de la API de Mercado Libre y de la creación de la tabla en la base de datos.
